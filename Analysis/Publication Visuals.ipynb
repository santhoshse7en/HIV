```python
from plotnine import *
import pandas as pd
```

```python
data = pd.read_csv('Final_HIV.csv').dropna().reset_index(drop=None)
data.head()
```

```python
sources = data.groupby('Publication').agg('count')
source_counts = sources.Articles.sort_values(ascending=False)
source_data = pd.DataFrame(source_counts).reset_index()
```

```python
sources_plot = ggplot(source_data, aes(x='Publication', y= 'Articles', fill='Publication')) + geom_col() + coord_flip()\
+ scale_fill_brewer(type='div', palette="Spectral") + theme_classic() + ggtitle('HIV articles published in different regions')\
+ geom_text(aes(label='Articles'), size=8, ha='right', format_string='{}')\

display(sources_plot)
```

```python
ggsave(plot=sources_plot, filename='publication.png', height=5, width=5, units = 'in', dpi=1000)
```
